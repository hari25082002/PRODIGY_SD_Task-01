#                     Task-02 Build a Temperature Conversion Program

```python
import tkinter as tk
from tkinter import ttk

# Constants for temperature conversion
C_TO_F = 9/5
C_TO_K = 273.15
F_TO_C = 5/9
F_TO_K = 273.15
K_TO_C = -273.15
K_TO_F = 9/5

def convert_temperature():
    try:
        input_temperature = float(Input.get())
    except ValueError:
        label.config(text="Invalid input. Please enter a valid number.")
        return

    selected_temperature = conversionlist.get()
    
    if selected_temperature == 'Celsius':
        Celsius = input_temperature
        Fahrenheit = Celsius * C_TO_F + 32
        Kelvin = Celsius + C_TO_K
    elif selected_temperature == 'Fahrenheit':
        Fahrenheit = input_temperature
        Celsius = (Fahrenheit - 32) * F_TO_C
        Kelvin = (Fahrenheit - 32) * F_TO_C + C_TO_K
    else:
        Kelvin = input_temperature
        Celsius = Kelvin + K_TO_C
        Fahrenheit = Kelvin * K_TO_F + 32
    
    label.config(text=f"Converted Temperatures:\nCelsius: {Celsius:.2f}\nFahrenheit: {Fahrenheit:.2f}\nKelvin: {Kelvin:.2f}")

T = tk.Tk()
T.title("Temperature_Converter")

# GUI elements
title_label = tk.Label(T, text="Temperature Converter", font=("Times New Roman", 18, 'bold'), background='#A9A9A9')
title_label.pack(padx=10)

main_frame = ttk.Frame(T, padding=(20, 10), style="Input.TFrame")
main_frame.pack()
input_label = ttk.Label(main_frame, text="Enter the Temperature:", font=("Arial", 14,), style="InputLabel.TLabel")
input_label.grid(row=0, column=0, padx=10, pady=10, sticky="e")
Input = ttk.Entry(main_frame, width=10, font=("Times New Roman", 12))
Input.grid(row=0, column=1, padx=10, pady=10)

# Set a default value for the conversion list
conversionlist = ttk.Combobox(main_frame, values=['Celsius', 'Fahrenheit', 'Kelvin'])
conversionlist.set('Celsius')
conversionlist.grid(row=0, column=2, padx=12, pady=12)

convert_button = ttk.Button(main_frame, text="Convert", command=convert_temperature, style="ConvertButton.TButton")
convert_button.grid(row=1, column=1, padx=15, pady=15)

result_frame = ttk.Frame(T, padding=(10, 20), style="Result.TFrame")
result_frame.pack()

label = ttk.Label(result_frame, text="", justify=tk.LEFT, font=("Arial", 14), style="ResultLabel.TLabel")
label.pack()

# Styling
style = ttk.Style()
style.configure("Input.TFrame", background="#F0F0F0")
style.configure("InputLabel.TLabel", background="#F0F0F0")
style.configure("ConvertButton.TButton", background="#4CAF50", foreground="black", font=("Arial", 14))
style.configure("Result.TFrame", background="#F0F0F0")
style.configure("ResultLabel.TLabel", background="#F0F0F0")

T.mainloop()
```
